## PARTE 1 ‚Äì Classe Principal e Objetos

**Tarefas:**
- Definir e criar a classe principal com pelo menos 3 atributos.
- Implementar o m√©todo construtor (`__init__`).

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 1
class SistemaReservas:
    def __init__(self, quantidade_maxima_salas):
        # Inicializando os atributos essenciais para o sistema
        self.salas = {}                # Dicion√°rio para armazenar as salas (n√∫mero da sala: objeto Sala)
        self.funcionarios = {}         # Dicion√°rio para armazenar os funcion√°rios (matr√≠cula: objeto Funcionario)
        self.reservas = []             # Lista para armazenar as reservas feitas
        self.autor = "Evelyn Forti"     # Nome do respons√°vel ou criador do sistema
        self.quantidade_maxima_salas = quantidade_maxima_salas  # Limite de salas

    def cadastrar_sala(self, numero, capacidade_maxima, tipo, recursos=None):
        if len(self.salas) >= self.quantidade_maxima_salas:
            print(f"‚ùå N√£o √© poss√≠vel cadastrar mais de {self.quantidade_maxima_salas} salas.")
            return
        
        if numero in self.salas:
            print(f"‚ùå Sala {numero} j√° cadastrada.")
            return

        self.salas[numero] = Sala(numero, capacidade_maxima, tipo, recursos)
        print(f"‚úÖ Sala {numero} cadastrada com sucesso.")

    def exibir_dados(self):
        print(f"\n===== Sistema de Reservas - Criado por {self.autor} =====")
        
        # Exibindo as informa√ß√µes das salas
        print("\nSalas Cadastradas:")
        if self.salas:
            for numero, sala in self.salas.items():
                print(sala)
        else:
            print("Nenhuma sala cadastrada ainda.")
        
        # Exibindo os funcion√°rios
        print("\nFuncion√°rios Cadastrados:")
        if self.funcionarios:
            for matricula, funcionario in self.funcionarios.items():
                print(funcionario)
        else:
            print("Nenhum funcion√°rio cadastrado ainda.")
        
        # Exibindo as reservas
        print("\nReservas Feitas:")
        if self.reservas:
            for reserva in self.reservas:
                print(f"Reserva de {reserva.funcionario.nome} - Sala: {reserva.numero_sala}, "
                      f"Data e Hora: {reserva.inicio} at√© {reserva.fim}, Pessoas: {reserva.quantidade_pessoas}")
        else:
            print("Nenhuma reserva realizada ainda.")

    def __str__(self):
        # M√©todo para representar a classe de forma leg√≠vel
        return f"Sistema de Reservas - Criado por {self.autor}. " \
               f"Atualmente com {len(self.salas)} salas (m√°ximo {self.quantidade_maxima_salas}), " \
               f"{len(self.funcionarios)} funcion√°rios e {len(self.reservas)} reservas registradas."


# Classe Sala (exemplo simples)
class Sala:
    def __init__(self, numero, capacidade_maxima, tipo, recursos=None):
        self.numero = numero
        self.capacidade_maxima = capacidade_maxima
        self.tipo = tipo
        self.recursos = recursos if recursos else []

    def __str__(self):
        recursos = ", ".join(self.recursos) if self.recursos else "Nenhum recurso"
        return f"Sala {self.numero} ({self.tipo}) - Capacidade M√°xima: {self.capacidade_maxima} - Recursos: {recursos}"


# Classe Reserva (exemplo simples)
class Reserva:
    def __init__(self, funcionario, numero_sala, inicio, fim, quantidade_pessoas):
        self.funcionario = funcionario
        self.numero_sala = numero_sala
        self.inicio = inicio
        self.fim = fim
        self.quantidade_pessoas = quantidade_pessoas


# Criando objetos de teste
sistema_escola = SistemaReservas(quantidade_maxima_salas=5)

# Cadastrar algumas salas
sistema_escola.cadastrar_sala(101, 30, "Sala de Aula", ["Projetor", "Computadores"])
sistema_escola.cadastrar_sala(102, 20, "Laborat√≥rio", ["Microsc√≥pio", "Computadores"])
sistema_escola.cadastrar_sala(103, 50, "Audit√≥rio", ["Microfone", "Palco"])

# Criando funcion√°rios e reservas
funcionario1 = type('Funcionario', (object,), {'nome': 'Jos√© Silva'})  # Exemplo de funcion√°rio
reserva1 = Reserva(funcionario1, 101, "2025-05-06 08:00", "2025-05-06 10:00", 25)
sistema_escola.reservas.append(reserva1)

# Exibir os dados do sistema
sistema_escola.exibir_dados()
```

```python
# Desenvolva aqui o c√≥digo da Parte 3
class SistemaReservas:  # Evelyn Forti
    def __init__(self, quantidade_maxima_salas):  # Evelyn Forti
        self.__salas = {}  # Evelyn Forti (atributo privado)  
        self.funcionarios = {}  # Evelyn Forti
        self.reservas = []  # Evelyn Forti
        self.__quantidade_maxima_salas = quantidade_maxima_salas  # Evelyn Forti
        self.autor = "Evelyn Forti"  # Evelyn Forti

    def get_quantidade_maxima_salas(self):  # Evelyn Forti
        return self.__quantidade_maxima_salas  # Evelyn Forti

    def set_quantidade_maxima_salas(self, nova_quantidade):  # Evelyn Forti
        if nova_quantidade >= len(self.__salas):  # Evelyn Forti
            self.__quantidade_maxima_salas = nova_quantidade  # Evelyn Forti
        else:  # Evelyn Forti
            print("‚ùå Nova quantidade √© menor que o n√∫mero de salas j√° cadastradas.")  # Evelyn Forti

    def cadastrar_sala(self, numero, capacidade_maxima, tipo, recursos=None):  # Evelyn Forti
        if len(self.__salas) >= self.__quantidade_maxima_salas:  # Evelyn Forti
            print("‚ùå Limite de salas atingido.")  # Evelyn Forti
            return  # Evelyn Forti
        if numero in self.__salas:  # Evelyn Forti
            print("‚ùå Sala j√° cadastrada.")  # Evelyn Forti
            return  # Evelyn Forti
        self.__salas[numero] = {"capacidade": capacidade_maxima, "tipo": tipo, "recursos": recursos or []}  # Evelyn Forti
        print(f"‚úÖ Sala {numero} cadastrada com sucesso!")  # Evelyn Forti

    def verificar_disponibilidade_sala(self, numero_sala):  # Evelyn Forti
        if numero_sala not in self.__salas:  # Evelyn Forti
            print("‚ùå Sala n√£o existe.")  # Evelyn Forti
            return False  # Evelyn Forti
        for reserva in self.reservas:  # Evelyn Forti
            if reserva["numero_sala"] == numero_sala:  # Evelyn Forti
                print("‚ùå Sala j√° reservada.")  # Evelyn Forti
                return False  # Evelyn Forti
        print("‚úÖ Sala dispon√≠vel.")  # Evelyn Forti
        return True  # Evelyn Forti

    def listar_salas(self):  # Evelyn Forti
        for numero, dados in self.__salas.items():  # Evelyn Forti
            print(f"Sala {numero} - Tipo: {dados['tipo']} - Capacidade: {dados['capacidade']}")  # Evelyn Forti


# Teste r√°pido da parte 3  # Evelyn Forti
sistema = SistemaReservas(2)  # Evelyn Forti
sistema.cadastrar_sala(101, 40, "Laborat√≥rio", ["Projetor"])  # Evelyn Forti
sistema.cadastrar_sala(102, 20, "Reuni√£o")  # Evelyn Forti
sistema.listar_salas()  # Evelyn Forti
sistema.verificar_disponibilidade_sala(101)  # Evelyn Forti
sistema.set_quantidade_maxima_salas(3)  # Evelyn Forti
print("M√°ximo de salas:", sistema.get_quantidade_maxima_salas())  # Evelyn Forti
```

```python
# Desenvolva aqui o c√≥digo da Parte 5
# Projeto Final: Sistema de Reservas - Evelyn Forti
import mysql.connector
from datetime import datetime

class BancoDeDados:
    def __init__(self, host='localhost', user='root', password='', database='reservas_db'):
        self.conn = mysql.connector.connect(
            host=host,
            user=user,
            password=password,
            database=database
        )
        self.cursor = self.conn.cursor()
        self._criar_tabela()

    def _criar_tabela(self):
        self.cursor.execute(
            """
            CREATE TABLE IF NOT EXISTS reservas (
                id INT AUTO_INCREMENT PRIMARY KEY,
                funcionario VARCHAR(100),
                numero_sala INT,
                inicio DATETIME,
                fim DATETIME,
                quantidade_pessoas INT
            )
            """
        )
        self.conn.commit()

    def inserir_reserva(self, funcionario, numero_sala, inicio, fim, quantidade_pessoas):
        self.cursor.execute(
            """
            INSERT INTO reservas (funcionario, numero_sala, inicio, fim, quantidade_pessoas)
            VALUES (%s, %s, %s, %s, %s)
            """,
            (funcionario, numero_sala, inicio, fim, quantidade_pessoas)
        )
        self.conn.commit()
        reserva_id = self.cursor.lastrowid
        print("‚úÖ Reserva inserida no banco.")
        return reserva_id

    def listar_reservas(self):
        self.cursor.execute("SELECT * FROM reservas")
        for row in self.cursor.fetchall():
            print(row)

    def excluir_reserva(self, reserva_id):
        self.cursor.execute("DELETE FROM reservas WHERE id = %s", (reserva_id,))
        self.conn.commit()
        print("‚ùå Reserva exclu√≠da do banco.")

    def sala_disponivel(self, numero_sala, inicio, fim):
        self.cursor.execute(
            """
            SELECT COUNT(*) FROM reservas
            WHERE numero_sala = %s
              AND (%s < fim AND %s > inicio)
            """,
            (numero_sala, inicio, fim)
        )
        return self.cursor.fetchone()[0] == 0

    def fechar_conexao(self):
        self.conn.close()

```
